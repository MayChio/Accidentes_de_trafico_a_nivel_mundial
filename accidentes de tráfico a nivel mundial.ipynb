```python
import pandas as pd

# Ruta completa al archivo CSV
file_path = r"C:\Users\MAYRA ROCIO\Downloads\Accidentes de trafico\road_accident_dataset.csv"  # Cambia el nombre del archivo si es necesario

# Cargar el archivo CSV
data = pd.read_csv(file_path)

# Ver las primeras filas del conjunto de datos
print(data.head())
```

```python
# Información básica sobre el dataset
print("\nInformación general del dataset:")
print(data.info())

# Estadísticas descriptivas del conjunto de datos
print("\nEstadísticas descriptivas del dataset:")
print(data.describe())

# Comprobación de valores nulos
print("\nValores nulos en cada columna:")
print(data.isnull().sum())

# Comprobación de columnas duplicadas
print("\nColumnas duplicadas:")
print(data.duplicated().sum())
```

```python
import matplotlib.pyplot as plt

# Distribución de accidentes por país
country_counts = data['Country'].value_counts()

# Mostrar las 10 principales países con más accidentes
plt.figure(figsize=(12, 6))
country_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 países con más accidentes de tráfico')
plt.xlabel('País')
plt.ylabel('Número de accidentes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribución de la gravedad de los accidentes
severity_counts = data['Accident Severity'].value_counts()

plt.figure(figsize=(8, 6))
severity_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'lightblue'])
plt.title('Distribución de la gravedad de los accidentes')
plt.ylabel('')  # Sin etiqueta para el eje y
plt.show()
```

```python
# Filtrar para visualizar la relación entre edad y nivel de alcohol
age_alcohol_relation = data[['Driver Age Group', 'Driver Alcohol Level']]

# Promediar el nivel de alcohol por grupo de edad
alcohol_by_age = age_alcohol_relation.groupby('Driver Age Group').mean()

# Visualizar la relación
plt.figure(figsize=(10, 6))
alcohol_by_age.plot(kind='bar', color='darkorange')
plt.title('Nivel de alcohol por grupo de edad del conductor')
plt.xlabel('Grupo de edad')
plt.ylabel('Nivel de alcohol')
plt.xticks(rotation=45)
plt.show()
```

```python
# Número de accidentes por condiciones meteorológicas
weather_counts = data['Weather Conditions'].value_counts()

# Mostrar las condiciones meteorológicas más comunes
plt.figure(figsize=(12, 6))
weather_counts.plot(kind='bar', color='lightblue')
plt.title('Número de accidentes por condiciones meteorológicas')
plt.xlabel('Condiciones meteorológicas')
plt.ylabel('Número de accidentes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gráfico de dispersión entre número de vehículos involucrados y número de fatalidades
plt.figure(figsize=(10, 6))
plt.scatter(data['Number of Vehicles Involved'], data['Number of Fatalities'], alpha=0.5, color='red')
plt.title('Relación entre número de vehículos involucrados y fatalidades')
plt.xlabel('Número de vehículos involucrados')
plt.ylabel('Número de fatalidades')
plt.show()
```
